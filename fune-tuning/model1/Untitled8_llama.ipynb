```python
from datasets import load_dataset
import pandas as pd
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import random
from datasets import load_dataset
from transformers import AutoConfig, AutoModelForCausalLM, BitsAndBytesConfig
import torch
```

```python
dataset = load_dataset("nvidia/OpenMathInstruct-2" ,
                       split="train[:4000]")
```

```python

# Load dataset
split = load_dataset("imdb")  
# Access train data
train_data = split["train"]

# Select 5 random samples
sample_indices = random.sample(range(len(train_data)), 5)

# Display samples
print("Random sample of 5 rows from train_data:")
for i, idx in enumerate(sample_indices):
    print(f"\n--- Sample {i+1} ---")
    print(train_data[idx]["text"])
```

```python
print("Dataset columns:", dataset.column_names)
```

## processing


```python
# Convert dataset directly from existing columns
def format_llama_example(example):
    return {
        "text": f"<|user|>\n{example['problem']}</s>\n<|assistant|>\n{example['generated_solution']}</s>"
    }

# Apply formatting + remove old columns
dataset = dataset.map(
    format_llama_example,
    remove_columns=dataset.column_names
)

# Split dataset
split = dataset.train_test_split(test_size=0.1, seed=42)

train_data = split["train"]
val_data = split["test"]

# Print results
print(f"Train: {len(train_data)} | Val: {len(val_data)}")
print(train_data[0]["text"][:300])
```

## cleaning

```python
num_null_values = sum(1 for example in train_data if example['text'] is None)
num_empty_strings = sum(1 for example in train_data if example['text'] is not None and len(example['text'].strip()) == 0)

print(f"Number of None values in 'text' column: {num_null_values}")
print(f"Number of empty strings (or strings with only whitespace) in 'text' column: {num_empty_strings}")

if num_null_values == 0 and num_empty_strings == 0:
    print("No null values or empty strings found in the 'text' column of train_data.")
else:
    print("Found null values or empty strings in the 'text' column of train_data.")
```

```python
# Convert the Hugging Face dataset to a pandas DataFrame for easier duplicate checking
train_df = pd.DataFrame(train_data)

# Check for duplicates in the 'text' column
duplicates = train_df[train_df.duplicated(subset=['text'], keep=False)]

if not duplicates.empty:
    print(f"Found {len(duplicates)} duplicate entries in the training data.")
    print("Here are some examples of duplicate entries:")
    for i, (index, row) in enumerate(duplicates.head(5).iterrows()):
        print(f"\n--- Duplicate Example {i+1} ---")
        print(row['text'])
else:
    print("No duplicate entries found in the training data.")
```

## Load Model

```python
MODEL_ID = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
```

```python
# load tokenizer
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
tokenizer.pad_token = tokenizer.eos_token
```

```python


config = AutoConfig.from_pretrained(MODEL_ID)

config.pad_token_id = config.eos_token_id

model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    config=config,
    quantization_config=BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_compute_dtype=torch.bfloat16,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4"
    ),
    device_map="auto"
)
print("Model loaded!")
```

```python
from peft import LoraConfig, prepare_model_for_kbit_training
from trl import SFTTrainer
from transformers import TrainingArguments

model = prepare_model_for_kbit_training(model)

trainer = SFTTrainer(
    model=model,
    train_dataset=train_data,
    eval_dataset=val_data,
    peft_config=LoraConfig(
        r=16,
        lora_alpha=32,
        target_modules=["q_proj", "v_proj", "k_proj", "o_proj"],
        lora_dropout=0.05,
        bias="none",
        task_type="CAUSAL_LM"
    ),
    args=TrainingArguments(
        output_dir="./llama-math",
        num_train_epochs=1,
        per_device_train_batch_size=2,
        gradient_accumulation_steps=8,
        learning_rate=2e-4,
        bf16=True,
        fp16=False,
        logging_steps=50,
        eval_strategy="steps",
        eval_steps=300,
        save_steps=300,
        report_to="none"
    ),
)

trainer.train()
print("Training done!")
```

## save model to s3

```python
import boto3, os
from google.colab import userdata

# Save model locally
trainer.save_model("./final-model")
tokenizer.save_pretrained("./final-model")

# Connect to S3
os.environ["AWS_ACCESS_KEY_ID"]     = userdata.get("AWS_ACCESS_KEY_ID")
os.environ["AWS_SECRET_ACCESS_KEY"] = userdata.get("AWS_SECRET_ACCESS_KEY")
os.environ["AWS_DEFAULT_REGION"]    = "us-east-1"

s3 = boto3.client("s3")

# Upload all files
for filename in os.listdir("./final-model"):
    s3.upload_file(
        f"./final-model/{filename}",
        "cloud-project-time4",
        f"models/llama-finetuned/{filename}"
    )
    print(f"Uploaded: {filename}")

print("All done! Model is on S3.")
```
